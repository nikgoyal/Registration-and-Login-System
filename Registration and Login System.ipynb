```python
from tkinter import *
import tkinter as tk
from tkinter.ttk import *
from tkinter import messagebox
import mysql.connector
from mysql.connector import Error
import re
import passlib.pwd  

root = tk.Tk()
root.geometry("300x250")
root.resizable(False,False)
def openNewWindow():
     
    newWindow = Toplevel(root)
    newWindow.title("New registration")
    newWindow.geometry("400x350")
    newWindow.resizable(False,False)
    label = Label(newWindow, text ="Registration Details")
    label.pack(pady = 10)
    
    def insert_data():
        id=entry_id.get()
        username=entry_name.get()
        age=entry_age.get()
        email=entry_email.get()
        password=entry_password.get()
        
        email_condition="^[a-z]+[\._]?[a-z 0-9]+[@]\w+[.]\w{2,3}$"
        has_digit= False
        
        if(id=="" and username=="" and age=="" and email=="" and password==""):
            messagebox.showerror("Error","You have to fill the form")

        elif re.search(email_condition,email) is None : 
            messagebox.showinfo("Error","Enter a valid email!")

        elif len(password)<8 :
            messagebox.showerror("Error","Enter a strong password!")

        else:
            
        
            try:
                #connect to mysql database
                conn= mysql.connector.connect(
                    host='localhost',
                    user='YOUR USERNAME',
                    password='YOUR PASSWORD',
                    database='YOUR DATABASE NAME'
                )
                if conn.is_connected():
                    cursor= conn.cursor()
        
                    #insert data
                    sql= "insert into YOUR TABLE NAME (id,username,age,email,password) values(%s,%s,%s,%s,%s)"
                    val=(id,username,age,email,password)
                    cursor.execute(sql,val)
                    conn.commit()
                    messagebox.showinfo("Success","Data inserted successfully!")
                    reset_fields()
                else:
                    messagebox.showerror("Error","Failed to connect to the database.")
            except Error as e:
                messagebox.showerror("Error",f"Error:{e}")
            finally:
                if conn.is_connected():
                    cursor.close()
                    conn.close()

        

    label_id=tk.Label(newWindow,text="ID:")
    label_id.place(x=40,y=70)
    
    entry_id=tk.Entry(newWindow)
    entry_id.place(x=120,y=70)
    
    label_name=tk.Label(newWindow,text="Username:")
    label_name.place(x=40,y=100)
    
    entry_name=tk.Entry(newWindow)
    entry_name.place(x=120,y=100)    
    

    label_age=tk.Label(newWindow,text="Age:")
    label_age.place(x=40,y=130)
    
    entry_age=tk.Entry(newWindow)
    entry_age.place(x=120,y=130)

    label_email=tk.Label(newWindow,text="Email:")
    label_email.place(x=40,y=160)
    
    entry_email=tk.Entry(newWindow)
    entry_email.place(x=120,y=160)
    
    label_password=tk.Label(newWindow,text="Password:")
    label_password.place(x=40,y=190)
    
    entry_password= tk.Entry(newWindow,show="*")
    entry_password.place(x=120,y=190)
    
    button_insert=tk.Button(newWindow,text="Register",command=insert_data)
    button_insert.place(x=120,y=250)

    def reset_fields():
        entry_id.delete(0, END)
        entry_name.delete(0, END)
        entry_age.delete(0, END)
        entry_email.delete(0,END)
        entry_password.delete(0, END)

    btn= Button(newWindow, text= "Quit", command= newWindow.destroy).place(x=250,y=270)


def openViewWindow():
    viewWindow = Toplevel(root)
    viewWindow.title("View Details")
    viewWindow.geometry("400x350")
    viewWindow.resizable(False,False)
    label = Label(viewWindow, text ="User Details")
    label.pack(pady = 10)
    
    conn= mysql.connector.connect(
        host='localhost',
        user='YOUR USERNAME',
        password='YOUR PASSWORD',
        database='YOUR DATABASE NAME'
    )
    if conn.is_connected():
        cursor= conn.cursor()
        conn.commit()
        # Create a cursor object to execute SQL commands
        
        cursor.execute("SELECT * from YOUR TABLE NAME")
        YOUR TABLE NAME = cursor.fetchall()
        listbox = Listbox(viewWindow)
    
        for user in YOUR TABLE NAME:
            ID = user[0]
            Name = user[1]
            
            listbox.insert(END, f"{ID}: {Name}")
        listbox.pack()
    
        conn.close()
        btn= Button(viewWindow, text= "Quit", command= viewWindow.destroy).place(x=250,y=270)
    else:
        messagebox.showerror("Error","Failed to connect to the database.")

def openLoginWindow():
    LoginWindow = Toplevel(root)
    LoginWindow.title("Login Details")
    LoginWindow.geometry("400x350")
    LoginWindow.resizable(False,False)
    label = Label(LoginWindow, text ="Login details")
    label.pack(pady = 10)

    def login_function():

        
        if(entry_name.get()=="" and entry_password.get()==""):
            messagebox.showerror("Error","You have to fill the form",parent=LoginWindow)
        else:
            try:
                conn= mysql.connector.connect(
                    host='localhost',
                    user='Your USERNAME',
                    password='YOUR PASSWORD',
                    database='YOUR DATABASE NAME'
                )
                cursor= conn.cursor()
                cursor.execute("select * from YOUR TABLE NAME where username=%s and password=%s",
                               (entry_name.get(),entry_password.get()))
                row=cursor.fetchone()
                if row == None:
                    messagebox.showerror("Error!","Invalid USERNAME & PASSWORD",parent=LoginWindow)
                else:
                    messagebox.showinfo("Success","Wellcome to the Project",parent=LoginWindow)
                    
                    conn.close()

            except Exception as e:
                messagebox.showerror("Error!",f"Error due to {str(e)}",parent=LoginWindow)
                    
            
    label_name=tk.Label(LoginWindow,text="Username:")
    label_name.place(x=40,y=100)

    entry_name=tk.Entry(LoginWindow)
    entry_name.place(x=120,y=100)

    label_password=tk.Label(LoginWindow,text="Password:")
    label_password.place(x=40,y=130)
    
    entry_password=tk.Entry(LoginWindow)
    entry_password.place(x=120,y=130)
    
    button_insert=tk.Button(LoginWindow,text="Login",command=lambda:[login_function(),LoginWindow.destroy()])
    button_insert.place(x=120,y=230)

    
    btn= Button(LoginWindow, text= "Quit", command= LoginWindow.destroy).place(x=250,y=270)
    
   
label = Label(root, text ="Login and registration")
label.pack(pady = 10)

btn = Button(root, text ="New registration", command = openNewWindow)
btn.pack(pady = 10)
btn1 = Button(root, text ="View Users", command = openViewWindow)
btn1.pack(pady = 10)
btn2= Button(root,text="Login User",command=openLoginWindow)
btn2.pack(pady=10)

Button(root, text= "Quit", command= root.destroy).place(x=150,y=200)
root.mainloop()
```
